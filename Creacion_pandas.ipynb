```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788'
df_toronto = pd.read_html(url)
```

```python
df_toronto[0:2]
```

```python
df_toronto[0]
```

```python
df_toronto = pd.DataFrame(df_toronto[0])
df_toronto.head()
```

```python
df_toronto.shape
```

```python
df_toronto = df_toronto[df_toronto['Borough'] != 'Not assigned']
df_toronto.head()
```

```python
df_toronto['Borough'].unique()
```

```python
df_toronto[df_toronto['Neighbourhood'] == 'Not assigned']
```

```python
len(df_toronto["Neighbourhood"][df_toronto["Neighbourhood"] == 'Not assigned']) 
```

```python
df_toronto = pd.DataFrame(df_toronto.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(lambda hoods: ", ".join(hoods)))
```

```python
df_toronto.reset_index(inplace = True)
df_toronto.head()
```

```python
df_toronto.shape
```

```python
df_toronto.rename(columns = {"Postal Code": "PostalCode"}, inplace= True)
df_toronto.head()
```
